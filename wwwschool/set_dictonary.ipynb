# Sets

### A set is a collection which is unordered, unchangeable*, and unindexed.

```python
this={'w','e','o','p','u','n'}
```

```python
print(this)
```

```python
print("p" in this)
```

```python
for i in this:
    print(i,end=' ')
```

```python
this.add("uyu")

print(this)
```

## update

#### to add items from another set into the current set, use the update() method.

```python
x={'l','t','z','r'}

this.update(x)

print(this)
```

```python
print(x)
```

### Remove

```python
# print(x.remove('t'))

this.remove('l')
print(this)
```

```python
this.clear()
print(this)
```

```python
this={'u', 'uyu', 'r', 'o', 'n', 'z', 't', 'e', 'p', 'w'}

del this
print(this)
```
