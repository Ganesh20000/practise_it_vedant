# TUPLE & LIST

## List

```python
L=['A','B','C','D','E','F','G','H']
```

##  Access List Items

```python
print(L)
print(L[1:5])
print(L[:5])


print('A' in L)

print( 'E' is L)
```

## Change List Items

```python
L[1:3]='O','U'

print(L)
```

### Change List Items

```python
print(L)



L.insert( 5,"Mongo")
print(L)
```

#### append

```python
L.append('aee')
print(L)
```

### Extend 

```python
L.extend(['rerer',"diva","ope"])

print(L)
```

#### REMOVE

```python
L.remove("aee")
print(L)
```

#### Pop

```python
L.pop(8)
print(L)
```

#### Del

```python
del L[4]
del L[0]

print(L)
```

```python
print(L.clear())
```

```python
L=["apple", "banana", "cherry"]
```

```python
for i in L:
    print(i,end=' ')
```

```python
for i in range(len(L)):
    print(L[i],end=' ')
```

# Tuple

```python
x=('p','i','o','e','p','v','y')

y=list(x)
print(y)

y.append("z")

x=tuple(y)

print(x)
```

#### loop through tuple

```python
for i in range(len(x)):
    print(x[i],end=' ')

print()
print("*"*50)
for i in x:
    print(i,end=' ')
```
